### A1. FizzBuzz

```python
def fizzbuzz(n):
    result = []
    for i in range(1, n + 1):
        if i % 15 == 0:
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 == 0:
            result.append("Buzz")
        else:
            result.append(str(i))
    return result

result = fizzbuzz(15)
print(result)
assert result[2] == "Fizz"
assert result[4] == "Buzz"
assert result[14] == "FizzBuzz"
assert result[0] == "1"
print("All assertions passed!")
```

### A2. Word frequency counter

```python
def word_freq(text):
    counts = {}
    for word in text.lower().split():
        word = word.strip(".,!?;:")
        counts[word] = counts.get(word, 0) + 1
    return counts

test = "The quick brown fox jumps over the lazy dog. The dog barked!"
freq = word_freq(test)
print(freq)
assert freq["the"] == 3
assert freq["dog"] == 2
print("All assertions passed!")
```

### A4. Matrix class

```python
class Matrix:
    def __init__(self, data):
        self.data = data
        self.rows = len(data)
        self.cols = len(data[0])

    def transpose(self):
        new_data = [[self.data[r][c] for r in range(self.rows)] for c in range(self.cols)]
        return Matrix(new_data)

    def __add__(self, other):
        if self.rows != other.rows or self.cols != other.cols:
            raise ValueError(f"Shape mismatch: ({self.rows},{self.cols}) vs ({other.rows},{other.cols})")
        new_data = [[self.data[r][c] + other.data[r][c] for c in range(self.cols)] for r in range(self.rows)]
        return Matrix(new_data)

    def __repr__(self):
        return "\n".join(str(row) for row in self.data)

A = Matrix([[1, 2, 3], [4, 5, 6]])
print("A:"); print(A); print(f"Shape: {A.rows}×{A.cols}")
print("\nA transposed:"); print(A.transpose())
B = Matrix([[10, 20, 30], [40, 50, 60]])
print("\nA + B:"); print(A + B)
```

### B1. Activation functions

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

x = np.linspace(-5, 5, 200)
funcs = [("Sigmoid", sigmoid), ("Tanh", tanh), ("ReLU", relu), ("Leaky ReLU", leaky_relu)]
colors = ["#378ADD", "#7F77DD", "#1D9E75", "#D85A30"]

fig, axes = plt.subplots(1, 4, figsize=(16, 3.5))
for ax, (name, fn), color in zip(axes, funcs, colors):
    ax.plot(x, fn(x), color=color, linewidth=2)
    ax.axhline(0, color="gray", linewidth=0.5)
    ax.axvline(0, color="gray", linewidth=0.5)
    ax.set_title(name)
    ax.set_xlabel("x")
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### B3. Vectorised distance matrix

```python
def pairwise_distances(points):
    # points shape: (n, d)
    # diff[i,j] = points[i] - points[j] for all pairs
    diff = points[:, np.newaxis, :] - points[np.newaxis, :, :]  # (n, n, d)
    return np.sqrt((diff ** 2).sum(axis=-1))  # (n, n)

pts = np.array([[0, 0], [3, 0], [0, 4]])
D = pairwise_distances(pts)
print(D)
# [[0. 3. 4.]
#  [3. 0. 5.]
#  [4. 5. 0.]]
assert np.allclose(D, D.T), "Should be symmetric"
assert np.allclose(np.diag(D), 0), "Diagonal should be zero"
print("All checks passed!")
```

### B4. Mini neural network forward pass

```python
np.random.seed(42)

def softmax(z):
    exp_z = np.exp(z - z.max(axis=1, keepdims=True))  # subtract max for numerical stability
    return exp_z / exp_z.sum(axis=1, keepdims=True)

# Input: 5 samples, 4 features
X = np.random.randn(5, 4)

# Initialise weights (small random values)
W1 = np.random.randn(4, 8) * 0.1
b1 = np.zeros(8)
W2 = np.random.randn(8, 3) * 0.1
b2 = np.zeros(3)

# Forward pass
hidden = np.maximum(0, X @ W1 + b1)  # ReLU
output = softmax(hidden @ W2 + b2)

print("Output (probabilities):")
print(output.round(4))
print("\nRow sums (should all be 1.0):", output.sum(axis=1).round(6))
print("Predicted classes:", output.argmax(axis=1))
```

### B5. Gradient descent on a 2D function

```python
def f(x, y):
    return (x - 3)**2 + 2*(y + 1)**2

def grad_f(x, y):
    return 2*(x - 3), 4*(y + 1)

# Gradient descent
lr = 0.1
x, y = 0.0, 0.0
path = [(x, y)]
values = [f(x, y)]

for _ in range(50):
    dx, dy = grad_f(x, y)
    x -= lr * dx
    y -= lr * dy
    path.append((x, y))
    values.append(f(x, y))

path = np.array(path)
print(f"Final: x={x:.4f}, y={y:.4f}, f={f(x,y):.6f}")

# Contour plot with path
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

xx, yy = np.meshgrid(np.linspace(-1, 5, 100), np.linspace(-3, 2, 100))
zz = f(xx, yy)
axes[0].contour(xx, yy, zz, levels=30, cmap="RdBu_r", alpha=0.7)
axes[0].plot(path[:, 0], path[:, 1], "o-", color="#D85A30", markersize=3, linewidth=1.5, label="GD path")
axes[0].plot(3, -1, "*", color="#1D9E75", markersize=15, label="Minimum (3, -1)")
axes[0].set_xlabel("x"); axes[0].set_ylabel("y")
axes[0].set_title("Gradient descent on f(x,y)")
axes[0].legend()

# Convergence
axes[1].plot(values, color="#378ADD", linewidth=2)
axes[1].set_xlabel("Step"); axes[1].set_ylabel("f(x, y)")
axes[1].set_title("Convergence")
axes[1].set_yscale("log")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Section C — Pandas

```python
# Recreate the dataset
np.random.seed(42)
n = 300
students = pd.DataFrame({
    "student_id": [f"S{i:03d}" for i in range(n)],
    "name": [f"Student_{i}" for i in range(n)],
    "department": np.random.choice(["CS", "EE", "Math", "Physics", "Biology"], n,
                                    p=[0.3, 0.25, 0.2, 0.15, 0.1]),
    "year": np.random.choice([1, 2, 3, 4], n, p=[0.3, 0.3, 0.25, 0.15]),
    "gpa": np.clip(np.random.normal(3.0, 0.5, n), 1.0, 4.0).round(2),
    "math_score": np.clip(np.random.normal(70, 15, n), 0, 100).round(1),
    "programming_score": np.clip(np.random.normal(65, 20, n), 0, 100).round(1),
    "statistics_score": np.clip(np.random.normal(68, 18, n), 0, 100).round(1),
    "study_hours_per_week": np.clip(np.random.exponential(12, n), 1, 50).round(1),
})
for col in ["math_score", "programming_score", "statistics_score", "study_hours_per_week"]:
    mask = np.random.rand(n) < 0.08
    students.loc[mask, col] = np.nan
```

### C1. Data cleaning & exploration

```python
# a) Missing values
print("Missing values:")
missing = students.isnull().sum()
pct = (missing / len(students) * 100).round(1)
print(pd.DataFrame({"count": missing, "pct": pct})[missing > 0])

# b) Fill scores with department medians
for col in ["math_score", "programming_score", "statistics_score"]:
    students[col] = students.groupby("department")[col].transform(
        lambda x: x.fillna(x.median())
    )

# c) Fill study hours with year means
students["study_hours_per_week"] = students.groupby("year")["study_hours_per_week"].transform(
    lambda x: x.fillna(x.mean())
)

# d) Verify
print("\nRemaining missing:", students.isnull().sum().sum())

# e) Average score
students["average_score"] = students[["math_score", "programming_score", "statistics_score"]].mean(axis=1).round(1)
print("\nSample with average_score:")
print(students[["name", "math_score", "programming_score", "statistics_score", "average_score"]].head())
```

### C2. Groupby analysis

```python
# a) Highest average GPA by department
gpa_by_dept = students.groupby("department")["gpa"].mean().sort_values(ascending=False)
print("Average GPA by department:")
print(gpa_by_dept.round(3))
print(f"Highest: {gpa_by_dept.index[0]}")

# b) Students per department per year
print("\nStudents per dept per year:")
print(students.groupby(["department", "year"]).size().unstack(fill_value=0))

# c) Correlation: study hours vs average score per department
print("\nCorrelation (study hours vs avg score):")
corr_by_dept = students.groupby("department").apply(
    lambda g: g["study_hours_per_week"].corr(g["average_score"]),
    include_groups=False
)
print(corr_by_dept.round(3))

# d) Top 3 students per department
print("\nTop 3 per department:")
top3 = students.groupby("department").apply(
    lambda g: g.nlargest(3, "average_score")[["name", "department", "average_score"]],
    include_groups=False
)
print(top3.to_string())
```

### C3. Feature engineering

```python
# a) Performance tier
def assign_tier(score):
    if score >= 85:
        return "Excellent"
    elif score >= 70:
        return "Good"
    elif score >= 55:
        return "Average"
    else:
        return "Needs improvement"

students["performance_tier"] = students["average_score"].apply(assign_tier)

# b) Score standard deviation
students["score_std"] = students[["math_score", "programming_score", "statistics_score"]].std(axis=1).round(2)

# c) Above department average
dept_means = students.groupby("department")["average_score"].transform("mean")
students["above_dept_avg"] = (students["average_score"] > dept_means).astype(int)

# d) Frequency table
print("Performance tier counts:")
print(students["performance_tier"].value_counts())

# e) Mean score_std per tier
print("\nMean score variability per tier:")
print(students.groupby("performance_tier")["score_std"].mean().round(2))
```

```python
np.random.seed(42)
epochs = np.arange(1, 101)
lrs = [0.001, 0.01, 0.1]
configs = [
    {"decay": 0.02, "floor": 0.4, "noise": 0.03, "val_gap": 0.15},
    {"decay": 0.06, "floor": 0.1, "noise": 0.02, "val_gap": 0.08},
    {"decay": 0.10, "floor": 0.2, "noise": 0.15, "val_gap": 0.25},
]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, lr, cfg in zip(axes, lrs, configs):
    train = 2.5 * np.exp(-cfg["decay"] * epochs) + cfg["floor"] + np.random.randn(100) * cfg["noise"]
    val = train + cfg["val_gap"] + np.random.randn(100) * cfg["noise"] * 1.5
    ax.plot(epochs, train, color="#378ADD", linewidth=1.5, label="Train")
    ax.plot(epochs, val, color="#D85A30", linewidth=1.5, label="Val")
    best_val = val.min()
    ax.axhline(best_val, color="gray", linestyle="--", linewidth=0.8, alpha=0.6, label=f"Best val={best_val:.2f}")
    ax.set_title(f"LR = {lr}")
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Loss")
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### E2. EDA report function

```python
def eda_report(df):
    print("=" * 50)
    print("EDA REPORT")
    print("=" * 50)
    print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns\n")

    print("Column types:")
    print(df.dtypes.to_string())

    missing = df.isnull().sum()
    if missing.sum() > 0:
        print(f"\nMissing values ({missing.sum()} total):")
        print(missing[missing > 0].to_string())
    else:
        print("\nNo missing values.")

    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

    if numeric_cols:
        print(f"\nNumeric columns ({len(numeric_cols)}):")
        print(df[numeric_cols].describe().round(2).to_string())

    # Plotting
    n_hist = len(numeric_cols)
    n_bar = len(cat_cols)
    has_corr = n_hist >= 2
    total_plots = n_hist + n_bar + (1 if has_corr else 0)
    if total_plots == 0:
        return

    cols = 3
    rows = (total_plots + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 4 * rows))
    axes = np.array(axes).flatten() if total_plots > 1 else [axes]

    idx = 0
    for col in numeric_cols:
        axes[idx].hist(df[col].dropna(), bins=20, color="#7F77DD", edgecolor="white", alpha=0.85)
        axes[idx].set_title(col)
        idx += 1

    if has_corr:
        sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap="RdBu_r",
                    center=0, ax=axes[idx], square=True, annot_kws={"size": 8})
        axes[idx].set_title("Correlations")
        idx += 1

    for col in cat_cols:
        vc = df[col].value_counts().head(10)
        axes[idx].barh(vc.index.astype(str), vc.values, color="#1D9E75")
        axes[idx].set_title(col)
        idx += 1

    for i in range(idx, len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()

# Test
eda_report(students)
```
